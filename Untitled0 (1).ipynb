```python
#read excel file
df = pd.read_excel('drive/My Drive/Internship_data/Baby_list_all.xlsx',header=[0])
df
```

```python
#restrict baby data
x = df['scan_age']
baby_41 = df[(x < 41) & (x >= 40)]
baby_41
```

```python
#plot first graph
import matplotlib.pylab as plt
scan_age = baby_41['birth_weight'].values
plt.figure(figsize=(5,5))
plt.hist(scan_age,color='r',bins = 18)
plt.title('Birth Weight Distribution')
plt.xlabel('Birth Weight (in kg)')
plt.ylabel('Frequency')
#plt.savefig('drive/My Drive/Internship_data/birth_weight_dist.png',pad_inches=0,bbox_inches='tight')
```

```python
#define birth age
birth_age = baby_41['birth_age']
```

```python
#define vpreterm
vpreterm = baby_41[(birth_age < 32)]
vpreterm
```

```python
#define modpreterm
modpreterm = baby_41[(birth_age >= 32) & (birth_age < 37)]
modpreterm
```

```python
#define term
term = baby_41[(birth_age >= 37)]
term
```

```python
#define heights for birth age dist. in groups
term ['scan_age'].values
num_term = len(term ['scan_age'].values)
num_vpreterm = len(vpreterm ['scan_age'].values)
num_modpreterm = len(modpreterm ['scan_age'].values)
```

```python
#define heights for ave weights
import numpy as np
ave_weight_vpreterm = np.mean(vpreterm ['birth_weight'].values)
ave_weight_modpreterm = np.mean(modpreterm ['birth_weight'].values)
ave_weight_term = np.mean(term ['birth_weight'].values)
```

```python
#drop babies with missing data
dfnew = df[['scan_age', 'scan_head_circumference', 'birth_age', 'birth_weight']]
dfn = dfnew.dropna()
baby_411 = dfn[(x < 41) & (x >= 40)]
vpreterm1 = baby_411[(birth_age < 32)]
modpreterm1 = baby_411[(birth_age >= 32) & (birth_age < 37)]
term1 = baby_411[(birth_age >= 37)]
```

```python
#define heights for ave circ
ave_shc_vpreterm = np.mean(vpreterm1 ['scan_head_circumference'].values)
ave_shc_modpreterm = np.mean(modpreterm1 ['scan_head_circumference'].values)
ave_shc_term = np.mean(term1 ['scan_head_circumference'].values)
```

```python
#plot multi graph
import matplotlib.pylab as plt
plt.figure(figsize = (10,5))
fig, ax1 = plt.subplots(1, 1)
Height_bar1 = [num_vpreterm, num_modpreterm, num_term]
Height_bar2 = [ave_weight_vpreterm, ave_weight_modpreterm, ave_weight_term]
Height_bar3 = [ave_shc_vpreterm, ave_shc_modpreterm, ave_shc_term]
plt.bar([1, 2, 3],Height_bar1, width = 0.5, color = ['red', 'orange', 'blue'], label = ['Extremely/Very Preterm', 'Moderate/Late Preterm', 'Full-term'])
plt.bar([5, 6, 7],Height_bar2, width = 0.5, color = ['red', 'orange', 'blue'])
plt.errorbar([5, 6, 7], [ave_weight_vpreterm, ave_weight_modpreterm, ave_weight_term], yerr = [np.std(vpreterm ['birth_weight'].values), np.std(modpreterm ['birth_weight'].values), np.std(term ['birth_weight'].values)], fmt = 'o', color = 'black', markersize = 2)
plt.bar([9, 10, 11],Height_bar3, width = 0.5, color = ['red', 'orange', 'blue'])
plt.errorbar([9, 10, 11], [ave_shc_vpreterm, ave_shc_modpreterm, ave_shc_term], yerr = [np.std(vpreterm1 ['scan_head_circumference'].values), np.std(modpreterm1 ['scan_head_circumference'].values), np.std(term1 ['scan_head_circumference'].values)], fmt = 'o', color = 'black', markersize = 2)
plt.xlim(0,12)

#axis titles

x1 = np.arange(2, 11, 4)
squad = ['Birth Age', 'Birth Weight', 'Scan Head Circumference']
ax1.set_xticks(x1)
ax1.set_xticklabels(squad, minor=False, rotation = 0)
plt.legend()
#plt.savefig('drive/My Drive/Internship_data/multi_info_graph.png',pad_inches=0,bbox_inches='tight')
```
